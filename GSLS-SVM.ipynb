```python
from GSLS_SVM import GSLS_Regression
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
x = np.array(range(1, 5, 1), dtype=np.float64)
y = np.array([x for x in x ])
print(x, y)

svm = GSLS_Regression(lambda x, y: np.dot(x, y), 100, 3)

obj = svm.fit(x, y)

y_result = np.array([svm.calc(x) for x in x])
plt.plot(x, y_result)
plt.plot(x, y, '.')

plt.figure()
plt.plot(obj)
```

```python
svm = GSLS_Regression(lambda x, y: np.exp(-((x - y) ** 2 / 2 / 3.5 ** 2)), 3000, 10)
obj = svm.fit(x, y)
y_result = np.array([svm.calc(x) for x in x])
plt.plot(x, y_result)
plt.plot(x, y)
plt.plot(x, y, '.')

support_x = np.array([x[i] for i in svm.S])
support_y = np.array([y[i] for i in svm.S])

plt.plot(support_x, support_y, 'D')

plt.figure()
plt.plot(obj)
```
